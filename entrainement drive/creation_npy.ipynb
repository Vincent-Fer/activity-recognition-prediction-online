```python
src_dir_train = '/content/ntu_encoder/train/'
src_dir_test = '/content/ntu_encoder/test/'

def creer_np(path,name):
  print(name)
  tab = []
  lab = []
  for subdir in os.listdir(path):
    print(subdir)
    for file in os.listdir(path + subdir):
      img = cv2.imread(path + subdir + '/' + file)
      tab.append(img)
      lab.append(classes.index(subdir))
  np.save('/content/drive/MyDrive/Colab Notebooks/vgg16/numpy/foot_to_foot_25/'+ name + 'tab_25_40_foot_to_foot.npy',np.asarray(tab))
  np.save('/content/drive/MyDrive/Colab Notebooks/vgg16/numpy/foot_to_foot_25/'+ name + 'label_25_40_foot_to_foot.npy',np.asarray(lab))

creer_np(src_dir_train,'train')
creer_np(src_dir_test,'test')
```
